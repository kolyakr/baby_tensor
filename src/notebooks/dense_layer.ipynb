```python
import sys
import os

project_root = os.path.dirname(os.path.abspath('..')) 

if project_root not in sys.path:
    sys.path.append(project_root)

from src.NeuralNetwork import NeuralNetwork
from src.layers import DenseLayer
import numpy as np
```

```python
layers = [
  DenseLayer(
    output_dim = 3,
    activation_name="relu",
    name = "Dense 1"
  ),
  DenseLayer(
    output_dim = 10,
    activation_name="relu",
    name = "Dense 2"
  ),
  DenseLayer(
    output_dim = 5,
    activation_name="relu",
    name = "Dense 3"
  ),
]

nn = NeuralNetwork(
  layers = layers,
  loss_type="mse",
  optimizer_type="momentum",
  seed=1
)
```

```python
X_train = np.random.normal(size = (20, 1000))
y_train = np.random.normal(size = (5, 1000))

nn.build(X_train.shape[0])
```

```python
nn.train(X_train, y_train, learning_rate=0.01)
```
