```python
layer1 = DenseLayer(
  name="Dense 1",
  output_dim=5,
  activation_name="relu"
)

layer2 = DenseLayer(
  name="Dense 2",
  output_dim=5,
  activation_name="sigmoid"
)

layer3 = DenseLayer(
  name="Dense 3",
  output_dim=5,
  activation_name="linear"
)

x = np.array(
  [
    [1, 2, 8],
    [1, 5, 3],
    [0, 2, 4],
  ]
)

layers = [layer1, layer2, layer3]
input_dim = x.shape[1]

for layer in layers:
  layer.initialize_parameters(input_dim)
  input_dim = layer.output_dim
  
```

```python
A_prev = x

for layer in layers:
  A_prev = layer.forward_pass(A_prev)
```

```python
y_pred = A_prev
loss = np.mean(y_pred)

dL_dy_pred = y_pred - 0.03
dA = dL_dy_pred


for layer in layers[::-1]:
  dA = layer.backward_pass(dA)
```
